## Null Distributions

```r
set.seed(1)
```

```r
n <- 1000
#empty vector
dist <- vector("numeric", n)

for(i in seq(1:n)){
    dist[i] <- mean( sample(x, 5) )
}
```

```r
diff <- abs(dist - mean(x))
```

```r
length(which(diff > 1)) / n
```

```r
set.seed(1)
n <- 10000
#empty vector
dist <- vector("numeric", n)

for(i in seq(1:n)){
    dist[i] <- mean( sample(x, 5) )
}
diff <- abs(dist - mean(x))
length(which(diff > 1)) / n
```

```r
hist(dist)
```

```r
set.seed(1)
n <- 1000
#empty vector
dist <- vector("numeric", n)

for(i in seq(1:n)){
    dist[i] <- mean( sample(x, 50) )
}
diff <- abs(dist - mean(x))
length(which(diff > 1)) / n
```

```r
hist(dist)
```

```r
#dplyr not working with R jupyter
expVec <- filter(gapminder, year==1952) %>% select(lifeExp) %>% unlist
```

```r
mean(expVec <= 60)- mean(expVec <= 40)
```

```r
prop = function(q) {
  mean(x <= q)
}
```

```r
qs = seq(from=min(x), to=max(x), length=20)
```

```r
props = sapply(qs, prop)
plot(qs, props)

#lambda 
props = sapply(qs, function(q) mean(x <= q))
```
